### Imports

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
```

### Data Loading and Initial Investigation

```python
#Load the dataset and look at the info of the dataset
csv_path = Path("concrete_data.xls")
dataset = pd.read_excel(csv_path)
dataset.info()
```

```python
#Check for missing data
dataset.isnull().sum()
```

```python
#Look at mean and standard deviation of the features with in the dataset as well as other statistical measures
dataset.describe()
```

```python
#Investiage Sample data points in the dataset (first and last 5 rows)
display(dataset.head())
display(dataset.tail())
```

```python
#Histogram of the plots of the dataset with mean shown as a vertical line
dataset.hist(bins=50, figsize=(20,15))
for i, ax in enumerate(plt.gcf().axes):
    ax.axvline(dataset.iloc[:, i].mean(), color='r', linestyle='dashed', linewidth=2, label='Mean')
    ax.legend()
plt.show()
```
